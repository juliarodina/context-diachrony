```python
grouped = df.groupby(['INPUT:word', 'INPUT:new', 'INPUT:old', 'INPUT:type'], 
                       as_index=False).agg({'OUTPUT:result': lambda x: list(x)[:5]})
```

```python
grouped.columns = ['word', 'sent1', 'sent2', 'group', 'annotations']
```

```python
grouped['mean'] = grouped['annotations'].apply(lambda x: mean(x))
```

```python
grouped['annotator1'] = grouped['annotations'].apply(lambda x: x[0])
grouped['annotator2'] = grouped['annotations'].apply(lambda x: x[1])
grouped['annotator3'] = grouped['annotations'].apply(lambda x: x[2])
grouped['annotator4'] = grouped['annotations'].apply(lambda x: x[3])
grouped['annotator5'] = grouped['annotations'].apply(lambda x: x[4])
```

```python
grouped = grouped.drop('annotations', axis=1)
```
