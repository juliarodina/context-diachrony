```python
import numpy as np
import pandas as pd
```

```python
dataset0 = pd.read_csv('dataset_0_ANNOTATED.tsv', sep='\t')
dataset1 = pd.read_csv('dataset_1_ANNOTATED.tsv', sep='\t')
```

```python
dataset0 = dataset0[(dataset0.annotator1 != 0) & (dataset0.annotator2 != 0) & (dataset0.annotator3 != 0)
                    & (dataset0.annotator4 != 0) & (dataset0.annotator5 != 0)]
len(dataset0)
```

```python
dataset1 = dataset1[(dataset1.annotator1 != 0) & (dataset1.annotator2 != 0) & (dataset1.annotator3 != 0)
                    & (dataset1.annotator4 != 0) & (dataset1.annotator5 != 0)]
len(dataset1)
```

```python
dataset0.head()
```

```python
grouped = dataset0.groupby(['word', 'group'], 
                       as_index=False).agg({'mean': lambda x: np.around(np.mean(x), decimals=2)})
grouped
```

```python
grouped = grouped.reset_index().groupby(['word', 'group'])['mean'].aggregate('first').unstack()
grouped
```

```python
grouped['delta_later'] = np.around(grouped['LATER'] - grouped['EARLIER'], decimals=2)
```
