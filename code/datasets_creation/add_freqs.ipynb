```python
import pandas as pd
```

```python
df = pd.read_csv('alpha_values.tsv', sep='\t', index_col=False)
df.head()
```

```python
df0 = df[df.dataset == '1']
df0 = df0[df0.alpha > 0.4]
print(len(df0))
df0
```

```python
df_durel = pd.read_csv('dataset_1_DUREL.tsv', sep='\t', index_col=False)
df_durel.head()
```

```python
df_durel = df_durel.loc[df_durel['word'].isin(df0.name.tolist())]
df_durel = df_durel.reset_index(drop=True)
```

```python
df_durel
```

```python
corpora1 = pd.read_csv('../macro/corpora/soviet.csv', index_col='ID')
corpora1.head()
```

```python
corpora2 = pd.read_csv('../macro/corpora/post-soviet.csv', index_col='ID')
corpora2.head()
```

```python
df_raw = pd.read_csv('dataset_1.csv', index_col=False)
df_raw['word1'] = df_raw['WORD'].apply(lambda x: x.split('_')[0])
df_raw.head()
```

```python
df_raw = df_raw.loc[df_raw['word1'].isin(df0.name.tolist())]
df_raw = df_raw.reset_index(drop=True)
```

```python
df_raw
```

```python
def get_contexts(word, corpora):
    samples = set()
    
    for idx, lemmas in corpora[['LEMMAS']].itertuples():
        if word in lemmas.lower().split():
            samples.add(lemmas)
      
    return samples
```

```python
results = pd.DataFrame(columns=['word', 'frequency_presov/sov'])

for word in df_raw['WORD'].tolist():
    old_samples = get_contexts(word.lower(), corpora1)
    new_samples = get_contexts(word.lower(), corpora2)
    
    results.loc[len(results)] = [word.split('_')[0], str(len(old_samples))+'/'+str(len(new_samples))]
```

```python
results
```

```python
final = pd.merge(df_durel, results, on='word')
final
```

```python
final.to_csv('dataset_1_DUREL_filtered.tsv', sep='\t', index=False)
```
